```r
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,25,100)
y = np.sin(x)

# Plotting using Matplotlib
plt.plot(x, y)
plt.title('Wave Plot')
plt.xlabel('eeee')
plt.ylabel('zzzzz')
plt.show()
```

```r
import seaborn as sns
import pandas as pd

data = pd.DataFrame({'x': np.random.randn(100), 'y': np.random.randn(100)})

sns.scatterplot(data=data, x='x', y='y')
plt.title('Scatter Plot')
plt.show()
```

```r
library(ggplot2)

x <- seq(0, 10, length.out = 100)
y <- sin(x)

ggplot(data.frame(x = x, y = y), aes(x = x, y = y)) +
  geom_line() +
  labs(title = " Wave Plot", x = "X", y = "Y")
```

```r
# Install and load the lattice library if not already installed
if (!requireNamespace("lattice", quietly = TRUE)) {
  install.packages("lattice")
}
library(lattice)

# Generating sample data
x <- rnorm(100)
y <- rnorm(100)

# Creating a scatter plot using lattice
xyplot(y ~ x, data = data.frame(x = x, y = y),
       main = "Scatter Plot ",
       xlab = "X", ylab = "Y")
```
