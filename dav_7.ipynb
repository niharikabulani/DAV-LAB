Tokenization (Sentence & Word)

```r
import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus# sample text for performing tokenization
text = "The tragedy of 'Dido, Queen of Carthage,' was probably completed for the stage after that irreparable and incalculable loss to English letters by Thomas Nash, the worthiest English precursor of Swift in vivid, pure, and passionate prose."
# importing word_tokenize from nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
# Passing the string text into word tokenize for breaking the sentences
token = word_tokenize(text)
token
```

Frequency Distribution


```r
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
fdist = FreqDist(token)
fdist.plot()
plt.show()
```

Remove stopwords & punctuations


```r
import string
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words("english"))
filtered_tokens = []
punctuations = list(string.punctuation)
for tok in token:
  if tok not in punctuations and tok not in stop_words:
    filtered_tokens.append(tok)


filtered_tokens
```

stemming

```r
from nltk.stem import PorterStemmer
pst = PorterStemmer()
stem_token = []
for tok in filtered_tokens:
  stem_token.append(pst.stem(tok))

stem_token
```

Part of Speech tagging

```r
import nltk
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)

pos_tags = nltk.pos_tag(tokens)

print(pos_tags)
```
